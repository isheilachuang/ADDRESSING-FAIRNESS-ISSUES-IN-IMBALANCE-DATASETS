```python
import matplotlib.pyplot as plt
import numpy as np

# Data for SVM variants 
methods = ['SVM', 'Universum SVM', 'Twin SVM', 
           'Universum Twin SVM', 'QSVM', 'Universum QSVM', 'QTSVM', 'Universum QTSVM']

# Replace these values with your actual accuracy and F1 scores
Cat_dog = [0.87, 1.0, 1.0, 1.0, 0.89, 0.91, 1.0, 1.0]
DeFungi = [0.83, 0.85, 0.95, 0.94, 0.8, 0.78, 1.0, 1.0]
Real_waste = [0.94, 0.95, 1.0, 0.99, 0.99, 0.98, 1.0, 1.0]   
Sentence = [0.19, 0.57, 0.33, 0.37, 0.45, 0.42, 0.76, 0.73]
QSAR = [0.19, 0.59, 0.33, 0.35, 0.45, 0.44, 0.76, 0.73]


# Create line plot
plt.figure(figsize=(12, 6))  # Set figure size

x = np.arange(len(methods))  # Numeric indices for methods

plt.plot(x, Cat_dog, marker='o', label='Cat and Dog', linestyle='-', linewidth=2)
plt.plot(x, DeFungi, marker='s', label='DeFungi', linestyle='-', linewidth=2)
plt.plot(x, Real_waste, marker='^', label='Real Waste', linestyle='-', linewidth=2)
plt.plot(x, Sentence, marker='d', label='Sentiment Labelled Sentences', linestyle='-', linewidth=2)
plt.plot(x, QSAR, marker='*', label='QSAR androgen receptor', linestyle='-', linewidth=2)

# Add labels, title, and legend
plt.xticks(x, methods, rotation=45, ha='right')  # Rotate x-axis labels
plt.ylabel('Score')
plt.xlabel('SVM Variants')
plt.title('Recall of SVM Variants_Imbalamced Data')
plt.ylim(0.1, 1.1)  # Adjust y-axis range
plt.legend(loc='upper right')

# Show grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
